```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
from mpl_toolkits.mplot3d import Axes3D
# Define the given functions
def s(x):
    return 0.5 * x
def q(x, y):
    integrand = lambda z: np.exp(-1/8 * (z - x)**2) / np.sqrt(2 * np.pi)
    result, _ = integrate.quad(integrand, -np.inf, y)
    return result
# Define parameters and functions
x_range = np.arange(0.1, 2.1, 0.1)
y_range = np.arange(0.1, 2.1, 0.1)
# Define constant c2
c2 = 0.2
# Define the interval for c1
c1_range = np.arange(0, 0.04, 0.001)
# Initialize result arrays
values_1 = np.zeros((len(x_range), len(y_range)))
values_2 = np.zeros((len(x_range), len(y_range), len(c1_range)))
# Calculate results
for i, x in enumerate(x_range):
    for j, y in enumerate(y_range):
        values_1[i, j] = -q(x, y) * s(x) - c2 + s(x)
        values_2[i, j, :] = q(x, y) * s(x) - c1_range - 0.1 * x  # Set values_2 as an array
# Initialize minimum and maximum values
min_value_1 = float('inf')
max_value_1 = float('-inf')
min_value_2 = float('inf')
max_value_2 = float('-inf')
# Loop through x, y, and c1 values
for x in x_range:
    for y in y_range:
        for c1 in c1_range:
            value_1 = -q(x, y) * s(x) - c2 + s(x)
            value_2 = q(x, y) * s(x) - c1 - 0.1 * x
            # Update minimum and maximum values
            min_value_1 = min(min_value_1, value_1)
            max_value_1 = max(max_value_1, value_1)
            min_value_2 = min(min_value_2, value_2)
            max_value_2 = max(max_value_2, value_2)
# Output the minimum and maximum values of the results
print(f"Minimum value of -q(x, y)s(x) - c2 + s(x): {min_value_1}, Maximum value: {max_value_1}")
print(f"Minimum value of q(x, y)s(x) - c1 - u(x): {min_value_2}, Maximum value: {max_value_2}")
# Create 3D plots
fig = plt.figure(figsize=(12, 6))
# Plot surface for -q(x, y)s(x) - c2 + s(x)
ax1 = fig.add_subplot(121, projection='3d')
x, y = np.meshgrid(x_range, y_range)
surface1 = ax1.plot_surface(x, y, values_1.T, cmap='viridis')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('-q(x, y)s(x) - c2 + s(x)')
ax1.set_title('-q(x, y)s(x) - c2 + s(x) Surface')
# Add color bar
fig.colorbar(surface1, ax=ax1, shrink=0.5, aspect=10)
# Plot surface for q(x, y)s(x) - c1 - u(x)
ax2 = fig.add_subplot(122, projection='3d')
surface2 = ax2.plot_surface(x, y, values_2[:, :, 0].T, cmap='viridis')  # Use the first array of values_2
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_zlabel('q(x, y)s(x) - c1 - u(x)')
ax2.set_title('q(x, y)s(x) - c1 - u(x) Surface')
# Add color bar
fig.colorbar(surface2, ax=ax2, shrink=0.5, aspect=10)
plt.show()
```

```python
import numpy as np
from scipy import integrate
import matplotlib.pyplot as plt

# Define the new integrand function
def new_integrand(z, x):
    return np.exp(-1/8 * (z - x)**2)

# Define the new upper limit function for integration
def new_upper_limit(y, x):
    result, _ = integrate.quad(lambda z: new_integrand(z, x), -np.inf, y)
    return result / np.sqrt(2 * np.pi)

# Generate values for y
y_values = np.linspace(0.1, 2, 100)

# Choose a specific x value (modify as needed)
x_value = 1.156

# Calculate integral values corresponding to y_values
integral_values = [new_upper_limit(y, x_value) for y in y_values]

# Plot the graph
plt.plot(y_values, integral_values, label=r'$\frac{1}{\sqrt{2\pi}}\int_{-\infty}^{y} e^{-\frac{1}{8}(z-x)^2} \, dz$')
plt.xlabel('y')
plt.ylabel('Integral Value')
plt.title('Graph of the Integral')
plt.legend()
plt.grid(True)
plt.show()

# Output the range of Integral Values
print("Range of Integral Values:", np.min(integral_values), "to", np.max(integral_values))
```
